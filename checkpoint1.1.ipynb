In this problem, you'll create a temperature conversion program that converts Celsius to Fahrenheit. The program prompts the user for a temperature in Celsius, converts the input to a float, and then calculates the equivalent temperature in Fahrenheit using the standard conversion formula.

Your task is to implement the helper functions `celsius_to_float` and `celsius_to_fahrenheit` that handle converting the input and performing the calculation.

### Exo 7: Study Schedule (10 points)

Different times of day are better for different types of studying:

- **7:00 - 9:00**: Focus session (problem-solving)
- **14:00 - 16:00**: Review session (practice)
- **22:00 - 24:00**: Coding session (programming practice)

**Your Task:**

Write a program that:
1. Prompts for the current time in 24-hour format (`H:MM` or `HH:MM`)
2. Outputs the recommended study activity
3. Outputs `"Break time"` if it's not an optimal study time

**Requirements:**

Include a `convert()` function that converts time strings to decimal hours.
- Example: `"7:30"` → `7.5`

Time ranges are inclusive.

**Examples:**

- Input: `8:30` → Output: `Focus session`
- Input: `15:00` → Output: `Review session`
- Input: `23:00` → Output: `Coding session`
- Input: `12:00` → Output: `Break time`



````python
def main():
    current_time = input("What time is it? ")
    converted_time = convert(current_time)
    print("the decimal time is ", converted_time)

    if 7.0 <= converted_time <= 9.0:
        print("Focus session")
    elif 14.0 <= converted_time <= 16.0:
        print("Review session")
    elif 22.0 <= converted_time <= 24.0:
        print("Coding session")
    else:
        print("Break time")

def convert(time):
    hours, minutes = time.split(":")
    hours = int(hours)
    minutes = int(minutes)

    return hours + minutes / 60


if __name__ == "__main__":
    main()
````

### Exo 8: File Type Detector (10 points)

Write a program that identifies file types based on their extensions.

**Your Task:**

1. Prompt the user for a filename
2. Output the file category based on its extension (case-insensitive)

**File Categories:**

- `.py` → `Python script`
- `.java` → `Java program`
- `.c` → `C program`
- `.txt` → `Text file`
- `.pdf` → `PDF document`
- `.jpg` or `.png` → `Image file`
- `.mp3` → `Audio file`

If the file has a different extension or no extension, output: `Unknown file type`

**Examples:**

- Input: `hello.py` → Output: `Python script`
- Input: `data` → Output: `Unknown file type`

````python
# Your code here, add as many cells as necessary
def main():
    #use lower to get the filename in lowercase
    filename = input("Enter a file: ").lower()

    # Check if the file has an extension
    if "." not in filename:
        print("Unknown file type")
        return 0

    # Get the extension after the last '.'
    extension = filename.split(".")[-1]

    if extension == "py":
        print("Python script")
    elif extension == "java":
        print("Java program")
    elif extension == "c":
        print("C program")
    elif extension == "txt":
        print("Text file")
    elif extension == "pdf":
        print("PDF document")
    elif extension == "jpg" or extension == "png":
        print("Image file")
    elif extension == "mp3":
        print("Audio file")
    else:
        print("Unknown file type")


if __name__ == "__main__":
    main()
````

### Exo 9: camel to snake case (10 points)

We saw in the workshop that, in Python, variable names use snake_case (words separated by underscores, all lowercase), but many other languages use camelCase (first word lowercase, subsequent words capitalized).

Examples:
- camelCase: `firstName`, `preferredFirstName`
- snake_case: `first_name`, `preferred_first_name`

**Your Task:**

Write a program that:
1. Prompts the user for a variable name in camelCase
2. Outputs the corresponding name in snake_case

**Examples:**

- Input: `firstName` → Output: `first_name`
- Input: `preferredFirstName` → Output: `preferred_first_name`

**Hint:** Remember that strings are iterable

````python
# Your code here, add as many cells as necessary
camel = input("Enter a name in camelCase: ")

snake = ""

#concatenate the characters in snake case
for char in camel:
    if char.isupper():
        snake += "_" + char.lower()
    else:
        snake += char

print(snake)
````

### Exo 11: Algerian Phone Number Validator (10 points)

In Algeria, phone numbers follow specific rules:
- **Mobile numbers**: Start with `05`, `06`, or `07`, followed by 8 more digits (10 digits total)
- **Home numbers**: Start with `02`, followed by 7 more digits (9 digits total)

**Rules for valid Algerian phone numbers:**

1. Mobile: Exactly 10 digits, starts with `05`, `06`, or `07`
2. Home: Exactly 9 digits, starts with `02`
3. Must contain only digits (no spaces, dashes, or other characters)

**Your Task:**

Write a program that:
1. Prompts the user for a phone number
2. Outputs `Valid` if it meets the requirements
3. Outputs `Invalid` if it doesn't

Structure your program with a function `is_valid(number)` that returns `True` or `False`.


````python
def main():
    number = input("Phone number: ")
    if is_valid(number):
        print("Valid")
    else:
        print("Invalid")

def is_valid(number):
    #check if the number is all digits
    if not number.isdigit():
        return False
    
    # Mobile (10 digits, starts with 05, 06 or 07)
    if len(number) == 10 and (number.startswith("05") or number.startswith("06") or number.startswith("07")):
        return True
    
    # Home (9 digits, starts with 02)
    if len(number) == 9 and number.startswith("02"):
        return True
        
    # If none of the above conditions are met
    return False

if __name__ == "__main__":
    main()
````

## Bonus 3: Grade Book (File I/O)(3 points)

Create a simple grade management system that reads from and writes to files.

**Your Task:**

Write a program that:
1. Reads student names and grades from a file `grades.txt` (create it yourselves)
2. Calculates the average grade for each student
3. Writes the results to `results.txt`

**Input file format (`grades.txt`):**
```
Ahmed 85 90 78
Fatma 92 88 95
Ali 70 75 80
```

**Output file format (`results.txt`):**
```
Ahmed: 84.3
Fatma: 91.7
Ali: 75.0

````python
def main():
    # Read from grades.txt 
    with open("grades.txt", "r") as infile, open("results.txt", "w") as outfile:
    # Process each line
        for line in infile:
            line = line.strip()
            parts = line.split()
            
            name = parts[0]
            grades = [int(g) for g in parts[1:]]
    # Calculate average
            average = sum(grades) / len(grades)
    # Write to results.txt        
            outfile.write(f"{name}: {average:.2f}\n")
    print("done")


if __name__ == "__main__":
    main()
````

### Bonus 4: Word Frequency Counter (Dictionaries) (3 points)


Analyze text and count word frequencies using dictionaries.

**Your Task:**

Write a program that:
1. Prompts the user for a sentence
2. Counts how many times each word appears (case-insensitive)
3. Displays words sorted by frequency (most common first)

**Example:**

Input: `"Python is great and Python is fun"`

Output:
```
python: 2
is: 2
great: 1
and: 1
fun: 1
```

**Requirements:**
- Use a dictionary to store word counts
- Ignore case (convert to lowercase)
- Remove punctuation
- Sort by frequency (descending)

**Hint:** You can use sorted() with a custom key

````python
def main():
    text = input("Enter text: ")
    punctuation = ".,!?;:\"()[]{}<>"
    # 1. Clean text (lowercase, remove punctuation)
    text = text.lower()  
    for p in punctuation: 
        text = text.replace(p, "")
    # 2. Split into words
    words = text.split()
    # 3. Count frequencies using dictionary
    dictio = {}
    for word in words:
        if word in dictio:
            dictio[word] += 1
        else:
            dictio[word] = 1
    # 4. Sort and display
    sorted_words = sorted(dictio.items(), key=lambda item: item[1], reverse=True)
    for word, count in sorted_words:
        print(f"{word}: {count}")

if __name__ == "__main__":
    main()
````
